```python
#계단 함수의 그래프
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype = np.int32)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#시그모이드 함수 구현
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-6.0, 5.0, 0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
#ReLU 함수 구현
def relu(x):
    return np.maximum(0,x)

x=np.arange(-5.0, 5.0, 0.1)
y=relu(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```
